```python
import pandas
import matplotlib.pyplot as plt
import random
from PyTechnicalIndicators.Bulk import basic_indicators, moving_averages, strength_indicators, candle_indicators
from PyTechnicalIndicators.Chart_Patterns import trends, peaks, pits

# TODO: at the end of each of these add one to show the use of single

data = pandas.read_csv("orcl.csv", sep=',', index_col=0, parse_dates=True)
data.index.name = 'Date'
data['Typical Price'] = (data['High'] + data['Low'] + data['Close']) / 3

# Log
log_df = pandas.DataFrame(index=data.index)
log_df['log'] = basic_indicators.log(data['Typical Price'])

fig_log = plt.figure()
ax_log = fig_log.add_subplot()
ax_log.plot(log_df)
```

```python
log_diff_df = pandas.DataFrame(index=data.index)
log_diff_df['log diff'] = basic_indicators.log_diff(data['Typical Price'])

fig_logdiff = plt.figure()
ax_logdiff = fig_logdiff.add_subplot()
ax_logdiff.plot(log_diff_df)
```

```python
# median
median_df = pandas.DataFrame(index=data.index, data=data['Typical Price'], copy=True)
median_df['5 Period Median'] = basic_indicators.median(data['Typical Price'], 5, True)
median_df['10 Period Median'] = basic_indicators.median(data['Typical Price'], 1, True)
median_df['20 Period Median'] = basic_indicators.median(data['Typical Price'], 20, True)

# TODO: have last 5 in a table
print(median_df[-5:])
```

```python
# mean
mean_df = pandas.DataFrame(index=data.index, data=data['Typical Price'], copy=True)
mean_df['5 Period Mean'] = basic_indicators.mean(data['Typical Price'], 5, True)
mean_df['10 Period Mean'] = basic_indicators.mean(data['Typical Price'], 10, True)
mean_df['20 Period Mean'] = basic_indicators.mean(data['Typical Price'], 20, True)

# TODO: have last 5 in a table
print(mean_df[-5:])
```

```python
# stddev
stddev_df = pandas.DataFrame(index=data.index, data=data['Typical Price'], copy=True)
stddev_df['5 Period Std Dev'] = basic_indicators.stddev(data['Typical Price'], 5, True)
stddev_df['10 Period Std Dev'] = basic_indicators.stddev(data['Typical Price'], 10, True)
stddev_df['20 Period Std Dev'] = basic_indicators.stddev(data['Typical Price'], 20, True)

# TODO: have last 5 in a table
print(stddev_df[-5:])
```

```python
# Variance
variance_df = pandas.DataFrame(index=data.index, data=data['Typical Price'], copy=True)
variance_df['5 Period Var'] = basic_indicators.variance(data['Typical Price'], 5, True)
variance_df['10 Period Var'] = basic_indicators.variance(data['Typical Price'], 10, True)
variance_df['20 Period Var'] = basic_indicators.variance(data['Typical Price'], 20, True)

print(variance_df[-5:])
```

```python
# Moving Average
moving_averages_df = pandas.DataFrame(data=data['Typical Price'], index=data.index, copy=True)
moving_averages_df['5 day MA'] = moving_averages.moving_average(moving_averages_df['Typical Price'], 5, True)
moving_averages_df['10 day MA'] = moving_averages.moving_average(moving_averages_df['Typical Price'], 10, True)
moving_averages_df['20 day MA'] = moving_averages.moving_average(moving_averages_df['Typical Price'], 20, True)

fig_ma = plt.figure()
ax_ma = fig_ma.add_subplot()
ax_ma.plot(moving_averages_df)
```

```python
# Smoothed Moving Average
smoothed_moving_averages_df = pandas.DataFrame(data=data['Typical Price'], index=data.index, copy=True)
smoothed_moving_averages_df['5 day SMA'] = moving_averages.smoothed_moving_average(
    smoothed_moving_averages_df['Typical Price'],
    5,
    True
)
smoothed_moving_averages_df['10 day SMA'] = moving_averages.smoothed_moving_average(
    smoothed_moving_averages_df['Typical Price'],
    10,
    True
)
smoothed_moving_averages_df['20 day SMA'] = moving_averages.smoothed_moving_average(
    smoothed_moving_averages_df['Typical Price'],
    20,
    True
)

fig_sma = plt.figure()
ax_sma = fig_sma.add_subplot()
ax_sma.plot(smoothed_moving_averages_df)
```

```python
# Exponential Moving Average
exponential_moving_averages_df = pandas.DataFrame(data=data['Typical Price'], index=data.index, copy=True)
exponential_moving_averages_df['5 day EMA'] = moving_averages.exponential_moving_average(
    exponential_moving_averages_df['Typical Price'],
    5,
    True
)
exponential_moving_averages_df['10 day EMA'] = moving_averages.exponential_moving_average(
    exponential_moving_averages_df['Typical Price'],
    10,
    True
)
exponential_moving_averages_df['20 day EMA'] = moving_averages.exponential_moving_average(
    exponential_moving_averages_df['Typical Price'],
    20,
    True
)

fig_ema = plt.figure()
ax_ema = fig_ema.add_subplot()
ax_ema.plot(exponential_moving_averages_df)
```

```python
# Personalised Moving Average
# TODO: explain random alpha nomitaor and denominator
random_alpha_nominator = random.uniform(0, 2)
print(f'Random Alpha nominator: {random_alpha_nominator}')
random_alpha_denominator = random.random()
print(f'Random Alpha nominator: {random_alpha_denominator}')

personalised_moving_averages_df = pandas.DataFrame(data=data['Typical Price'], index=data.index, copy=True)
personalised_moving_averages_df['5 day PMA'] = moving_averages.personalised_moving_average(
    personalised_moving_averages_df['Typical Price'],
    5,
    random_alpha_nominator,
    random_alpha_denominator,
    True
)
personalised_moving_averages_df['10 day PMA'] = moving_averages.personalised_moving_average(
    personalised_moving_averages_df['Typical Price'],
    10,
    random_alpha_nominator,
    random_alpha_denominator,
    True)
personalised_moving_averages_df['20 day PMA'] = moving_averages.personalised_moving_average(
    personalised_moving_averages_df['Typical Price'],
    20,
    random_alpha_nominator,
    random_alpha_denominator,
    True)

fig_pma = plt.figure()
ax_pma = fig_pma.add_subplot()
ax_pma.plot(personalised_moving_averages_df)
```

```python
# MACD
macd = moving_averages.moving_average_convergence_divergence(data['Typical Price'])
signal_line = moving_averages.signal_line(macd)
macd_diff = [macd[i] - signal_line[i-9] for i in range(len(macd[9:]))]
macd_x = data.index[34:]

fig_macd = plt.figure()
ax_macd = fig_macd.add_subplot()
ax_macd.plot(macd_x, macd[9:])
ax_macd.plot(macd_x, signal_line)
ax_macd.bar(macd_x, macd_diff)
```

```python
# get peak trend
highs = list(data['High'])
# get peaks angle
peak_list = peaks.get_peaks(highs)
peak_angle = trends.get_trend_angle(peak_list[0][0], 0, peak_list[-1][0], len(peak_list))
print(peak_angle)

peak_trend = trends.get_peak_trend(highs)
print(peak_trend)

fig_peaks = plt.figure()
ax_peaks = fig_peaks.add_subplot()
ax_peaks.plot(data['High'])
for i in peak_list:
    ax_peaks.scatter(data.index[i[1]], i[0], c='Red', marker="o")
```

```python
#get pit trend
lows = list(data['Low'])
# get pits angle
pits_list = pits.get_pits(lows)
pits_angle = trends.get_trend_angle(pits_list[0][0], 0, pits_list[-1][0], len(peak_list))
print(pits_angle)

pit_trend = trends.get_pit_trend(lows)
print(pit_trend)

fig_pits = plt.figure()
ax_pits = fig_pits.add_subplot()
ax_pits.plot(data['Low'])
for i in pits_list:
    ax_pits.scatter(data.index[i[1]], i[0], c='Red', marker="o")
```
